```python
# 데이터 병합
pandemic_1 = pd.concat([df_2002, df_2003]).dropna().reset_index(drop=True)
pandemic_2 = df_2008.dropna().reset_index(drop=True)
pandemic_3 = pd.concat([df_2011, df_2012, df_2101]).dropna().reset_index(drop=True)
pandemic_4 = pd.concat([df_2107, df_2108, df_2109, df_2110])
pandemic_4 = pandemic_4.drop('index', axis=1).dropna().reset_index(drop=True)
pandemic_5 = pd.concat([df_2111, df_2112])
pandemic_5 = pandemic_5.drop('index', axis=1).dropna().reset_index(drop=True)

pandemic_1 = pandemic_1[['날짜', '제목', '내용', 'url']].dropna()
pandemic_2 = pandemic_2[['날짜', '제목', '내용', 'url']].dropna()
pandemic_3 = pandemic_3[['날짜', '제목', '내용', 'url']].dropna()
pandemic_4 = pandemic_4[['날짜', '제목', '내용', 'url']].dropna()
pandemic_5 = pandemic_5[['날짜', '제목', '내용', 'url']].dropna()
```

```python
from konlpy.tag import Mecab

mecab = Mecab()

text_prep = pandemic_1['내용'].str.replace("[^ㅏ-ㅣ가-힣. ]", " ")
pandemic_1['text_prep'] = text_prep

text_corpus = [mecab.nouns(corpus) for corpus in text_prep]
```

```python
# 데이터 전처리

from konlpy.tag import Mecab
mecab = Mecab()

def data_preprocessing(df):
    text_prep = df['내용'].str.replace("[^ㅏ-ㅣ가-힣. ]", " ")
    df['text_prep'] = text_prep
    
    text_corpus = [mecab.nouns(corpus) for corpus in text_prep]

    # 1글자 제거
    tokenized_text = []
    for token_text in text_corpus:
        token_text = [word for word in token_text if len(word) > 1]
        tokenized_text.append(token_text)

    df['tokenized_text'] = tokenized_text
    
    return df

pandemic_1 = data_preprocessing(pandemic_1)
pandemic_2 = data_preprocessing(pandemic_2)
pandemic_3 = data_preprocessing(pandemic_3)
pandemic_4 = data_preprocessing(pandemic_4)
pandemic_5 = data_preprocessing(pandemic_5)
```

```python
print(pandemic_1['tokenized_text'][:20])
```

```python
pandemic_1['jeju_place'] = ''
pandemic_2['jeju_place'] = ''
pandemic_3['jeju_place'] = ''
pandemic_4['jeju_place'] = ''
pandemic_5['jeju_place'] = ''

pandemic_1['jeju_place'] = [list() for s in pandemic_1['jeju_place']]
pandemic_2['jeju_place'] = [list() for s in pandemic_2['jeju_place']]
pandemic_3['jeju_place'] = [list() for s in pandemic_3['jeju_place']]
pandemic_4['jeju_place'] = [list() for s in pandemic_4['jeju_place']]
pandemic_5['jeju_place'] = [list() for s in pandemic_5['jeju_place']]
```

```python
for col in list(df['구분'].unique()):
    print(col)
```

```python
# '구분' 칼럼에 따른 장소명 변수 생성
i=1
for col in list(df['구분'].unique()):
    globals()['places_{}'.format(i)]= list(df[df['구분']==col]['장소명'].unique())
    i+=1
    
# 장소 확인 함수
def place_check(df, places): 
    for idx, token_list in enumerate(df['tokenized_text']):
        place_tmp = []
        for token in token_list:
            if token in places:
                if token not in place_tmp: # 중복 append 방지
                    place_tmp.append(token)

        df['jeju_place'].iloc[idx].extend(place_tmp)
    
    return df
```

```python
for i in range(len(list(df['구분'].unique()))):
    pandemic_1 = place_check(pandemic_1, globals()['places_{}'.format(i+1)])

for i in range(len(list(df['구분'].unique()))):
    pandemic_2 = place_check(pandemic_2, globals()['places_{}'.format(i+1)])

for i in range(len(list(df['구분'].unique()))):
    pandemic_3 = place_check(pandemic_3, globals()['places_{}'.format(i+1)])

for i in range(len(list(df['구분'].unique()))):
    pandemic_4 = place_check(pandemic_4, globals()['places_{}'.format(i+1)])

for i in range(len(list(df['구분'].unique()))):
    pandemic_5 = place_check(pandemic_5, globals()['places_{}'.format(i+1)])
```
