```python
# KoNLPY 객체 생성
okt = Okt()
kkma = Kkma()
komoran = Komoran()
hannanum = Hannanum()
```

```python
# Dataset Load
pandemic = pd.read_excel('N차대유행 합본.xlsx')
pandemic.head()
```

```python
# 관광명소별 데이터셋 분류
udo = pandemic.loc[pandemic['jeju_place_google'] == '우도']
sungsan = pandemic.loc[pandemic['jeju_place_google'] == '성산일출봉']
subji = pandemic.loc[pandemic['jeju_place_google'] == '섭지코지']
osulok = pandemic.loc[pandemic['jeju_place_google'] == '오설록티뮤지엄']
joongmoon = pandemic.loc[pandemic['jeju_place_google'] == '중문관광단지']
hanrim = pandemic.loc[pandemic['jeju_place_google'] == '한림공원']
shinhwa = pandemic.loc[pandemic['jeju_place_google'] == '신화테마파크']
hyupjae = pandemic.loc[pandemic['jeju_place_google'] == '협재해변']
dongmoon = pandemic.loc[pandemic['jeju_place_google'] == '제주동문시장']
```

```python
# Series -> String
udo_pos = udo['tokenized_text'].to_string()
sungsan_pos = sungsan['tokenized_text'].to_string()
subji_pos = subji['tokenized_text'].to_string()
osulok_pos = osulok['tokenized_text'].to_string()
joongmoon_pos = joongmoon['tokenized_text'].to_string()
hanrim_pos = hanrim['tokenized_text'].to_string()
shinhwa_pos = shinhwa['tokenized_text'].to_string()
hyupjae_pos = hyupjae['tokenized_text'].to_string()
dongmoon_pos = dongmoon['tokenized_text'].to_string()
```

```python
# Okt 형태소 분류기를 이용한 품사 태깅
udo_pos = okt.pos(udo_pos)
sungsan_pos = okt.pos(sungsan_pos)
subji_pos = okt.pos(subji_pos)
osulok_pos = okt.pos(osulok_pos)
joongmoon_pos = okt.pos(joongmoon_pos)
hanrim_pos = okt.pos(hanrim_pos)
shinhwa_pos = okt.pos(shinhwa_pos)
hyupjae_pos = okt.pos(hyupjae_pos)
dongmoon_pos = okt.pos(dongmoon_pos)
```

```python
# 관광명소별 품사태깅 데이터프레임 생성
udo_pos = pd.DataFrame(udo_pos)
udo_pos.rename(columns={0:'word', 1:'pos'}, inplace=True)

sungsan_pos = pd.DataFrame(sungsan_pos)
sungsan_pos.rename(columns={0:'word', 1:'pos'}, inplace=True)

subji_pos = pd.DataFrame(subji_pos)
subji_pos.rename(columns={0:'word', 1:'pos'}, inplace=True)

osulok_pos = pd.DataFrame(osulok_pos)
osulok_pos.rename(columns={0:'word', 1:'pos'}, inplace=True)

joongmoon_pos = pd.DataFrame(joongmoon_pos)
joongmoon_pos.rename(columns={0:'word', 1:'pos'}, inplace=True)

hanrim_pos = pd.DataFrame(hanrim_pos)
hanrim_pos.rename(columns={0:'word', 1:'pos'}, inplace=True)

shinhwa_pos = pd.DataFrame(shinhwa_pos)
shinhwa_pos.rename(columns={0:'word', 1:'pos'}, inplace=True)

hyupjae_pos = pd.DataFrame(hyupjae_pos)
hyupjae_pos.rename(columns={0:'word', 1:'pos'}, inplace=True)

dongmoon_pos = pd.DataFrame(dongmoon_pos)
dongmoon_pos.rename(columns={0:'word', 1:'pos'}, inplace=True)
```

```python
# Number, Punctuation, Josa 품사 제거
def pos_drop(df):
  df.drop(df.loc[df['pos']=='Number'].index, inplace=True)
  df.drop(df.loc[df['pos']=='Punctuation'].index, inplace=True)
  df.drop(df.loc[df['pos']=='Josa'].index, inplace=True)

pos_drop(udo_pos)
pos_drop(sungsan_pos)
pos_drop(subji_pos)
pos_drop(osulok_pos)
pos_drop(joongmoon_pos)
pos_drop(hanrim_pos)
pos_drop(shinhwa_pos)
pos_drop(hyupjae_pos)
pos_drop(dongmoon_pos)
```

```python
# 관광지 place 칼럼 추가 
udo_pos['place'] = '우도'
sungsan_pos['place'] = '성산일출봉' 
subji_pos['place'] = '섭지코지' 
osulok_pos['place'] = '오설록티뮤지엄' 
joongmoon_pos['place'] = '중문관광단지' 
hanrim_pos['place'] = '한림공원' 
shinhwa_pos['place'] = '신화테마파크' 
hyupjae_pos['place'] = '협재해변' 
dongmoon_pos['place'] = '제주동문시장' 
```
